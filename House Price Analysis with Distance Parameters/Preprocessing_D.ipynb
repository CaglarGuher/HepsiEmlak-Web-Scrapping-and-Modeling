```python
import pandas as pd
```

```python
df = pd.read_csv("house_prices_with_distances.csv")
```

```python
df.isna().sum()
```

```python
rent_column_index = df.columns.get_loc('Bulunduğu Kat')

columns_to_drop = df.columns[rent_column_index + 1:]

df = df.drop(columns=columns_to_drop)
```

```python
df.isna().sum()
```

```python
df.drop(columns=["Yapı Tipi","Cephe","Otogar","Yapının Durumu","Tapu Durumu","Kullanım Durumu","Yapının Durumu","Takas","Kira Getirisi","Yetkili Ofis","City","İlan no","Son Güncelleme Tarihi","Unnamed: 0","Bakkal/Market","Şarküteri","Zincir Marketler","Lise","İlkokul","Üniversite Kampüs"],inplace=True)
```

```python
df.isna().sum()
```

```python
df.drop(columns=["Yakıt Tipi","Aile Sağlığı Merkezi","Sağlık Kurumları/Diğer"],inplace=True)
```

```python
df
```

```python
df = df.dropna()
```

```python
df
```

```python
df.isna().sum()
```

```python
df['Bina Yaşı'] =df['Bina Yaşı'].astype(str)

df['Bina Yaşı'] = df['Bina Yaşı'].str.replace(r'\D+', '', regex=True)

df['Bina Yaşı'] = pd.to_numeric(df['Bina Yaşı'], errors='coerce').fillna(0)
```

```python
df['Kat Sayısı'] = df['Kat Sayısı'].str.extract('(\d+)').astype(float)
df['Rent'] = df['Rent'].str.replace('.', '').str.replace(' TL', '')
df['Rent'] = df['Rent'].astype(float)
df['Oda + Salon Sayısı'] = df['Oda + Salon Sayısı'].apply(lambda x: int(x.split('+')[0]) + int(x.split('+')[1]))

# Remove 'm2' from Area and convert to integers
df['Brüt / Net M2'] = df['Brüt / Net M2'].str.replace(' m2', '').astype(float)

df.dtypes
columns_to_process = df.columns[10:17]

# Iterate over the selected columns
for column in columns_to_process:
    df[column] = df[column].str.replace(',', '.', regex=True)

    # Remove 'm' and convert to float
    mask_m = df[column].astype(str).str.contains(' m')
    df.loc[mask_m, column] = df.loc[mask_m, column].str.replace(' m', '').astype(float) / 1000

    # Remove 'km', convert to float, and multiply by 1000
    mask_km = df[column].astype(str).str.contains(' km')
    df.loc[mask_km, column] = df.loc[mask_km, column].str.replace(' km', '').astype(float)
    df[column] *= 1000

```

```python
df
```

```python
df = df.rename(columns={
    'İlçe': 'District',
    'Mahalle': 'Neighborhood',
    'Oda + Salon Sayısı': 'Num_Of_Room',
    'Brüt / Net M2': 'Area',
    'Bina Yaşı': 'Age',
    'Isınma Tipi': 'Heat_type',
    'Kat Sayısı': 'Apt_Floor',
    'Krediye Uygunluk': 'Credit_Eli',
    'Eşya Durumu': 'Items',
    'Banyo Sayısı': 'Num_of_Bath',
    'Metro İstasyonu': 'Metro_Station',
    'Durak (Otobüs)': 'Bus_Stop',
    'Durak (Minibüs)': 'Minibus_Stop',
    'Hastane': 'Hospital',
    'Acil Servis': 'Emergency_Service',
    'Üniversite': 'University',
    'Alışveriş Merkezi': 'Shopping_Mall',
    'Rent': 'Price',
    'Bulunduğu Kat': 'Floor'
})
```

```python
df["Minibus_Stop"] =df["Minibus_Stop"]/1000
df["Metro_Station"] = df["Metro_Station"]/1000
df["Bus_Stop"] = df["Bus_Stop"]/1000
df["Hospital"]= df["Hospital"]/1000
df["Emergency_Service"] = df["Emergency_Service"]/1000
df["University"] = df["University"]/1000
df["Shopping_Mall"] =df["Shopping_Mall"]/1000
```

```python
column_to_move = 'Price'
column = df.pop(column_to_move)
df.insert(len(df.columns), column_to_move, column)
df.reset_index(drop=True, inplace=True)
```
