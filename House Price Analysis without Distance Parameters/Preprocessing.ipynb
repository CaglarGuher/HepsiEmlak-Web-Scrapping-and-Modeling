```python
import pandas as pd

df = pd.read_csv("house_prices_without_distances.csv", index_col=0)
```

```python
df =df.drop(columns=["İlan no", "Son Güncelleme Tarihi", "Takas"])
df.isna().sum()
```

```python
df = df.iloc[:,:18]
df = df.dropna()
df.isna().sum()
```

```python
df
```

```python
df = df.rename(columns={'Oda + Salon Sayısı': 'Num_Of_Room', 'Brüt / Net M2': 'Area',
                               'Bulunduğu Kat': 'Floor',"Bina Yaşı":"Age","Isınma Tipi":"Heat_type","Kat Sayısı":"Apt_Floor",
                               "Krediye Uygunluk":"Credit_Eli","Eşya Durumu":"Items","Banyo Sayısı":"Num_of_Bath","Kullanım Durumu":"Usage","Cephe":"Side","Rent":"Price"})
```

```python


columns_to_keep = ['District', 'Neighborhood', 'Num_Of_Room', 'Area', 'Floor', 'Age', 'Heat_type', 'Apt_Floor',
                   
                   'Credit_Eli', 'Items', 'Num_of_Bath', 'Usage', 'Price']
```

```python
df = df[columns_to_keep]
```

```python
df
```

```python
unique_values = df['Age'].unique()
value_counts = df['Age'].value_counts()

# Print unique values in the 'Age' column
print("Unique values in the 'Age' column:")
print(unique_values)

# Print count of each unique value in the 'Age' column
print("\nCount of each unique value in the 'Age' column:")
print(value_counts)
```

```python
df['Floor'] = df['Floor'].replace(["Zemin", "Giriş Katı", "Bahçe Katı", "Yüksek Giriş"], 'first floor')
df['Floor'] = df['Floor'].replace(["Kot 1", "Kot 2", "Kot 3", "Bodrum", "Bodrum ve Zemin", "Yarı Bodrum"], 'basement')
df['Floor'] = df['Floor'].replace(["En Üst Kat", "Teras Katı", "Çatı Katı"], 'top floor')
df = df[~df['Floor'].isin(["Asma Kat"])]
df['Floor'] = df['Floor'].replace(df['Floor'].unique(), 'interstage')
df['Age'] = df['Age'].str.extract('(\d+)').astype(float)
df['Heat_type'] = df['Heat_type'].replace(df['Heat_type'].value_counts()[-6:].index.tolist(), 'Other')
df['Heat_type'] = df['Heat_type'].replace(df['Heat_type'].value_counts()[-5:].index.tolist(), 'Other')
df['Apt_Floor'] = df['Apt_Floor'].str.extract('(\d+)').astype(float)
df['Credit_Eli'] = df['Credit_Eli'].replace(df['Credit_Eli'].value_counts()[-1:].index.tolist(), 'Other')
df['Items'] = df['Items'].replace({'Eşyalı Değil': 0, 'Eşyalı': 1})
df['Credit_Eli'] = df['Credit_Eli'].map({"Uygun": 1, "Uygun değil": 0})
df = df[~df['Usage'].isin(df['Usage'].value_counts()[-1:].index.tolist())]
df['Price'] = df['Price'].str.replace('.', '').str.replace(' TL', '').astype(float)
df['Num_Of_Room'] = df['Num_Of_Room'].apply(lambda x: int(x.split('+')[0]) + int(x.split('+')[1]))
df['Area'] = df['Area'].str.replace(' m2', '').astype(float)
df['Age'] = df['Age'].fillna(0).astype(float)
```

```python
column_to_move = 'Price'
column = df.pop(column_to_move)
df.insert(len(df.columns), column_to_move, column)
df.reset_index(drop=True, inplace=True)
```
